```python
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Sequential
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import seaborn as sns

```

### Dataset Preparation

```python

#dataset generate
circle = make_circles(n_samples=1000, noise=0.2, factor=0.2)
data = np.concatenate([circle[0],circle[1].reshape(-1,1)],axis=1)
df = pd.DataFrame(data, columns=['X1', 'X2','y'])
#train test split
X_train, X_test, y_train, y_test = train_test_split(df[['X1', 'X2']], df['y'], test_size=0.3, shuffle=True)

sns.scatterplot(df,x='X1', y='X2',hue='y')
```

### Model
#### Model build

```python
model = Sequential()
input_layer = Input(shape=(2,))
linear_1 = Dense(4, activation='relu')
linear_2 = Dense(4, activation='relu')
out = Dense(1, activation='sigmoid')

model.add(input_layer)
model.add(linear_1)
model.add(linear_2)
model.add(out)
```

#### Model compile


```python
model.compile(optimizer='sgd', loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50)
```

```python
model.evaluate(X_test, y_test)
```
